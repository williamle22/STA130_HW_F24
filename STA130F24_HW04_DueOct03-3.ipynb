The "standard deviation" of the original data measures the variability of individual data points around the mean, capturing how much the observations differ from one another within that specific dataset. This statistic is essential for understanding the spread and dispersion of the data, indicating whether the values are closely clustered around the mean or widely scattered. In contrast, the "standard error of the mean" (SEM) quantifies the extent to which the sample mean would vary if multiple samples were drawn from the same population. It reflects the uncertainty associated with estimating the population mean based on a particular sample and is crucial for determining the reliability of this estimate. The SEM is calculated by dividing the standard deviation of the sample by the square root of the sample size, highlighting that larger samples yield a more precise estimate of the population mean. Essentially, while standard deviation focuses on the variability within the dataset itself, the SEM provides insight into the reliability and stability of the sample mean as an estimate of the population mean. Furthermore, the SEM decreases with increasing sample size, meaning that larger samples result in greater confidence in the mean estimate, while the standard deviation remains unchanged regardless of sample size.

To create a 95% confidence interval using the standard error of the mean (SEM), begin by calculating the sample mean from your data. Next, determine the SEM by dividing the standard deviation of your sample by the square root of the sample size. For a 95% confidence interval, you typically use a critical value from the standard normal distribution, which is approximately 1.96 for large samples. The confidence interval is then constructed by taking the sample mean and adding and subtracting the product of the SEM and the critical value, represented mathematically as: Confidence Interval = Sample Mean ± (1.96 × SEM). This interval represents the range within which we expect 95% of the bootstrapped sample means to fall, reflecting our uncertainty about the population mean. Thus, this approach provides a meaningful estimate of the true mean based on the sample data and helps in understanding the reliability of our estimate.

To create a 95% bootstrapped confidence interval directly using the bootstrapped means, you first generate a large number of bootstrap samples from your original dataset and calculate the mean for each of these samples, resulting in a distribution of bootstrapped means. Once you have this distribution, you can determine the 95% confidence interval by identifying the 2.5th percentile and the 97.5th percentile of the bootstrapped means. This involves taking the values that fall at these percentiles to form the interval. This method captures the central 95% of the bootstrapped sample means, providing a direct and empirical way to estimate the confidence interval without relying on the standard error. This approach effectively reflects the variability in the data and gives a more accurate representation of the confidence interval based on the bootstrap distribution.s

````python
import numpy as np

# This is the function to perform bootstrap sampling and calculate a confidence interval
def bootstrap_ci(data, num_bootstrap_samples=1000, confidence_level=0.95, statistic=np.median):
    # This will create a list to hold the results of the bootstrap samples
    bootstrap_statistics = []
    
    # Now we actually perform bootstrap sampling
    for _ in range(num_bootstrap_samples):
        # Randomly sample from the original data with replacement (like the video showed)
        # 'replace=True' means we can select the same data point more than once (sampling with replacement)
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the median and standard deviation (anything but the mean) for the bootstrap sample
        bootstrap_statistics.append(statistic(bootstrap_sample))
    
    # Now we calculate the lower and upper bounds for the confidence interval
    # The confidence level is usually set to 0.95 for a 95% confidence interval
    # The 2.5th percentile gives us the lower bound while the 97.5th percentile gives us the upper bound
    lower_percentile = (1 - confidence_level) / 2 * 100  # 2.5th percentile
    upper_percentile = (1 + confidence_level) / 2 * 100  # 97.5th percentile
    
    # Use np.percentile to find the values at these percentiles from the bootstrap statistics
    lower_bound = np.percentile(bootstrap_statistics, lower_percentile)  # To find the lower bound
    upper_bound = np.percentile(bootstrap_statistics, upper_percentile)  # To find the upper bound
    
    return lower_bound, upper_bound  # Return the confidence interval bounds

# Sample data from the ChatBot (for testing the function)
sample_data = np.array([1.5, 2.5, 3.5, 4.5, 5.5]) 

# Calculate the 95% bootstrap confidence interval for the population median
median_ci = bootstrap_ci(sample_data, statistic=np.median)
print(f"95% Bootstrap Confidence Interval for the Median: {median_ci}")

# To calculate the 95% bootstrap confidence interval for the standard deviation:
def sample_std(sample):
    # Calculate the sample standard deviation (using ddof=1 for an unbiased estimate)
    # 'ddof=1' adjusts the divisor during standard deviation calculation for a sample
    return np.std(sample, ddof=1)

# Now actually calculate the confidence interval for the standard deviation
std_ci = bootstrap_ci(sample_data, statistic=sample_std)
print(f"95% Bootstrap Confidence Interval for the Standard Deviation: {std_ci}")
````

Distinguishing between the population parameter and the sample statistic is crucial when discussing confidence intervals because they represent different aspects of statistical inference. The population parameter is a fixed value that describes a characteristic of the entire population (e.g., the true mean or median), while the sample statistic is a value calculated from a sample drawn from that population (e.g., the sample mean or median). Confidence intervals are used to estimate the range within which the population parameter is likely to fall, based on the variability and uncertainty inherent in using a sample. By understanding this distinction, we acknowledge that our sample statistic is only an estimate and may not perfectly reflect the population parameter. Confidence intervals help quantify this uncertainty and provide a more reliable basis for making inferences about the population.

6.1 Imagine you’ve got a bag of marbles that represents a whole population. Instead of taking out every single marble to learn about the entire bag, bootstrapping is like taking a handful of marbles out, examining them, and then putting them back. Essentially, you take your sample of marbles (your data) and randomly pick marbles from that sample, allowing yourself to select the same marble multiple times. You do this repeatedly to create many new “samples.” Each of these samples provides insight into what the whole bag might look like. So, you’re essentially creating mini-bags of marbles from your original handful multiple times to see how different characteristics or properties might vary.

6.2 The main purpose of bootstrapping is to help us understand the uncertainty or variability in our estimates. Let’s say you learned something from your handful of marbles, but you want to know how reliable that information is. Bootstrapping allows you to create numerous different scenarios based on your repeated sampling, which gives you a range of possible outcomes. This process helps you create confidence intervals—like saying, “I am pretty sure the true characteristics of the whole bag fall somewhere within this range.” It’s like giving yourself a safety net when making predictions about the larger population, allowing you to make decisions with greater assurance!

6.3 Let’s say you think there are 10 marbles in the whole bag (your hypothesized guess), but you only have your handful (your sample) to rely on. Here’s how you can use bootstrapping: First, you take your sample and create mini-bags by randomly picking marbles from your sample, allowing the same marble to be picked more than once (this is called sampling with replacement). After you create several mini-bags, you examine the characteristics of each one. Next, you check how many of these mini-bags support your guess of 10 marbles. If many of the mini-bags suggest something close to 10, then your guess seems reasonable. However, if most of the mini-bags show something very different from 10, your guess might not be accurate. To take it further, you can find a 95% confidence interval from the mini-bags (the middle 95% of the values). You can do this by taking the 2.5th and 97.5th percentiles (the lower and upper bound of the confidence interval) and using Python code. This is the range of values where you are 95% sure the true number of marbles in the whole bag falls. If your guess of 10 marbles is inside this confidence interval, it means your guess is plausible. But if the guess of 10 is outside that range, it’s likely that your guess is incorrect. All in all, bootstrapping helps you evaluate how reasonable your guess is based on what you’ve observed in your sample!

A confidence interval overlapping zero indicates that the range of plausible values for the population mean includes zero, which corresponds to the null hypothesis stating that there is no effect on average. Even if the observed sample mean is not zero, the uncertainty associated with the sample mean—due to variability and sample size—suggests that we cannot confidently assert that the observed effect is significant. This situation would lead us to reject the null hypothesis, as it indicates sufficient evidence that the drug does have an average effect. Essentially, the decision to reject or fail to reject the null hypothesis hinges on whether the observed sample mean’s variability allows us to confidently assert that it differs from zero in a statistically meaningful way.

### Quantitative Analysis

In this analysis, we aimed to evaluate the effectiveness of a new vaccine developed by AliTech by comparing the InitialHealthScore and FinalHealthScore of individuals who received the vaccine. To do this, we first prepared the dataset by calculating the Difference between the FinalHealthScore and InitialHealthScore for each patient, allowing us to directly assess the change in health resulting from the vaccine. The calculated differences for each patient revealed a range of improvements and declines, leading to an overall mean difference of 3.3.

To evaluate the statistical significance of this mean difference, we employed the bootstrapping method, a resampling technique that estimates the sampling distribution of a statistic by repeatedly sampling with replacement from the observed data. This involved generating a large number of bootstrap samples (e.g., 1,000 samples) from the original dataset and calculating the mean difference of health scores for each sample, thus creating a distribution of mean differences.

Next, we computed the mean of the bootstrapped mean differences and derived a 95% confidence interval (CI) using the percentiles of this distribution. For our analysis, the confidence interval was found to be [0.7,5.5]. The null hypothesis was defined as the assumption that there is no effect of the vaccine, represented by a hypothesized mean difference of 0. By comparing this hypothesized mean to our computed confidence interval, we found that 0 is not included in the interval [0.7,5.5].

The purpose of employing bootstrapping in this analysis was twofold: first, it provides a robust method for estimating confidence intervals and making inferences without relying on assumptions about the underlying data distribution; second, it quantifies the variability in our mean difference estimates, which is crucial for understanding the uncertainty in the vaccine's effectiveness. This approach allows us to establish a reliable 95% confidence interval for the mean difference in health scores.
