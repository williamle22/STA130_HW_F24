Q1. In Simple Linear Regression (SLR), we model the relationship between two variables by fitting a straight line to the data. This line represents how the predictor variable $X$ (independent variable) influences the outcome variable $Y$ (dependent variable). The model equation is 

$Y$ = $\beta_0$ + $\beta_1$ $X$ + $\epsilon$ 

where $\beta_0$ is the intercept (the value of $Y$ when $X$ = 0 and $\beta_1$ is the slope coefficient (the change in $Y$ for a one-unit increase in $X$. The error term $\epsilon$ accounts for random variation in $Y$ that cannot be explained by $X$ alone, and it is assumed to follow a normal distribution with a mean of zero. 

In essence, the SLR model predicts $Y$ values around the line $Y$ = $\beta_0$ + $\beta_1$ $X$, with the actual data points scattered due to the random error term. This formulation implies that observed $Y$ values are samples from a normal distribution centered on the line defined by $\beta_0$ + $\beta_1$ $X$, with deviations caused by the error term $\epsilon$.

````python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define parameters
np.random.seed(42)
beta_0 = 3   # intercept
beta_1 = 1.5 # slope
sigma = 3    # increased standard deviation of the error term

# Generate synthetic predictor data
X = np.linspace(0, 10, 100)

# Generate error term (epsilon) from a normal distribution with higher variability
epsilon = norm.rvs(loc=0, scale=sigma, size=X.size)

# Generate outcome data using the SLR model equation with increased variability
Y = beta_0 + beta_1 * X + epsilon

# Plot the results
plt.scatter(X, Y, color="skyblue", label="Sample Data")
plt.plot(X, beta_0 + beta_1 * X, color="orange", label="True Regression Line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression Model")
plt.legend()
plt.show()
````

````python
# Q2.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy.stats import norm

# Step 1: Define parameters for the theoretical model
np.random.seed(42)
beta_0 = 3   # intercept
beta_1 = 1.5 # slope
sigma = 3    # standard deviation of the error term

# Step 2: Generate synthetic predictor data
X = np.linspace(0, 10, 100)

# Step 3: Generate error term (epsilon) from a normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=X.size)

# Step 4: Generate outcome data using the SLR model equation
Y = beta_0 + beta_1 * X + epsilon

# Step 5: Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 6: Fit a Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Step 7: Print the summary of the fitted model
print(model.summary())

# Step 8: Visualize the data and the fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color="skyblue", label="Sample Data")
plt.plot(data['X'], model.fittedvalues, color="orange", label="Fitted Regression Line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
````

Q3. In **Question 1**, the **true regression line** is a **deterministic** representation of the relationship between the predictor variable $X$ and the outcome variable $Y$. It is given by the equation $Y$ = $\beta_0$ + $\beta_1$$X$ + $\epsilon$, where $\beta_0$ and $\beta_1$ are fixed parameters, and $\epsilon$ represents the error term. The true line is deterministic, meaning that for any given value of $X$, the corresponding value of $Y$ is exactly determined, with no randomness or error. This line represents the perfect, error-free relationship between $X$ and $Y$, assuming the absence of noise or sampling variation.

In **Question 2**, the **fitted regression line** is derived from simulated data, which includes the random error term $\epsilon$. This error term introduces randomness into the observed data, causing the fitted regression line, which is estimated from the data, to deviate from the true line. The fitted line is an approximation of the underlying relationship, but it is influenced by the random variation in the data caused by the error term. Since the error term varies in each simulation, the fitted regression line changes slightly with each new dataset.

The key difference is that the **true regression line** is **deterministic** and fixed, representing the ideal, error-free relationship between $X$) and $Y$, while the **fitted regression line** is subject to random sampling variation. The fitted line fluctuates around the true line due to the random error in each dataset. This illustrates how **random sampling variation** affects the regression model’s ability to perfectly estimate the true relationship: because the error term varies with each random sample, the fitted regression line is always an approximation and can never perfectly match the true line. Over repeated simulations, the fitted line may get closer to the true line on average, but it will always show some degree of fluctuation due to the inherent randomness in the data.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy.stats import norm

# Step 1: Define parameters for the theoretical model
np.random.seed(42)
beta_0 = 3   # intercept
beta_1 = 1.5 # slope
sigma = 3    # standard deviation of the error term

# Step 2: Generate synthetic predictor data
X = np.linspace(0, 10, 100)

# Step 3: Generate error term (epsilon) from a normal distribution
epsilon = norm.rvs(loc=0, scale=sigma, size=X.size)

# Step 4: Generate outcome data using the SLR model equation
Y = beta_0 + beta_1 * X + epsilon

# Step 5: Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 6: Fit a Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Step 7: Print the summary of the fitted model
print(model.summary())

# Step 8: Visualize the data, true regression line, and fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color="skyblue", label="Sample Data")

# True Regression Line
plt.plot(X, beta_0 + beta_1 * X, color="orange", label="True Regression Line")

# Fitted Regression Line from the model
plt.plot(X, model.fittedvalues, color="green", label="Fitted Regression Line")

plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression Model: True vs. Fitted Lines")
plt.legend()
plt.show()
````

````python
# Q10.

import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Filter the dataset for long wait times (assuming long_wait is defined as waits > some threshold)
long_wait = old_faithful['waiting'] > 63  # Adjust threshold based on "long wait" definition
long_wait_data = old_faithful[long_wait]

# Bootstrap sampling to estimate the slope coefficients
bootstrapped_slope_coefficients = []

# Number of bootstrap samples
n_bootstraps = 1000

for _ in range(n_bootstraps):
    # Generate a bootstrap sample
    sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    
    # Fit a simple linear regression model
    model = smf.ols('duration ~ waiting', data=sample).fit()
    
    # Collect the slope coefficient explicitly by name
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Convert to numpy array for easier calculations
bootstrapped_slope_coefficients = np.array(bootstrapped_slope_coefficients)

# Plot the bootstrapped slope coefficient distribution
sns.histplot(bootstrapped_slope_coefficients, kde=True)
plt.xlabel("Slope Coefficient")
plt.title("Bootstrapped Sampling Distribution of the Slope Coefficients")
plt.show()

# Simulate data under the null hypothesis
beta_0 = 1.65
beta_1 = 0
sigma = 0.37

simulated_slope_coefficients = []

for _ in range(n_bootstraps):
    # Create a simulated dataset with no association
    old_faithful_simulation = long_wait_data.copy()
    old_faithful_simulation['duration'] = beta_0 + beta_1 * old_faithful_simulation['waiting'] + stats.norm(loc=0, scale=sigma).rvs(size=len(old_faithful_simulation))
    
    # Fit a linear regression model to the simulated dataset
    model_sim = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    
    # Collect the slope coefficient explicitly by name
    simulated_slope_coefficients.append(model_sim.params['waiting'])

# Convert to numpy array for easier calculations
simulated_slope_coefficients = np.array(simulated_slope_coefficients)

# Plot the simulated slope coefficient distribution
sns.histplot(simulated_slope_coefficients, kde=True)
plt.xlabel("Slope Coefficient")
plt.title("Simulated Sampling Distribution of the Slope Coefficients Under Null Hypothesis")
plt.show()

# Calculate 95% confidence interval
ci_lower, ci_upper = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print(f"95% Bootstrapped Confidence Interval for Slope: ({ci_lower}, {ci_upper})")

# Check if 0 is within the confidence interval
contains_zero = (ci_lower <= 0 <= ci_upper)
print("Does the 95% CI contain 0?", contains_zero)

# Observed slope from the original data
observed_slope = smf.ols('duration ~ waiting', data=long_wait_data).fit().params['waiting']

# Calculate the simulated p-value
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print(f"Simulated p-value: {simulated_p_value}")

# Original model p-value
original_model_p_value = smf.ols('duration ~ waiting', data=long_wait_data).fit().pvalues['waiting']
print(f"Original model p-value: {original_model_p_value}")
````

### 13. The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 11, explaining how they're similar and different.<br>

> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_

````python
# Q13.
# A

import numpy as np

# Observed difference in means between short and long wait times
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test
n_permutations = 1000
permutation_diffs = []

for _ in range(n_permutations):
    # Shuffle the 'kind' labels
    shuffled_data = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1, replace=False).values)
    # Calculate the mean difference with shuffled labels
    shuffled_diff = shuffled_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(shuffled_diff)

# Calculate p-value (two-sided)
p_value = (np.abs(permutation_diffs) >= np.abs(observed_diff)).mean()
print(f"Observed Difference in Means: {observed_diff}")
print(f"P-value from Permutation Test: {p_value}")
````

````python
# B

bootstrapped_diffs = []

# Number of bootstrap samples
n_bootstraps = 1000

for _ in range(n_bootstraps):
    # Bootstrap sampling within each group
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    # Calculate the mean difference
    bootstrapped_diff = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_diffs.append(bootstrapped_diff)

# Calculate 95% confidence interval
conf_interval = np.quantile(bootstrapped_diffs, [0.025, 0.975])
print(f"95% Bootstrap Confidence Interval for Difference in Means: {conf_interval}")
````

(a) **Explanation of the Sampling Approaches**:
- **Permutation Test**: This test assesses the null hypothesis of no difference in mean `duration` between "short" and "long" wait times by shuffling the `kind` labels (short vs. long) across the dataset. For each permutation, the difference in mean `duration` between the two shuffled groups is calculated, creating a distribution of differences expected under the null hypothesis. The observed difference is then compared to this distribution, and the p-value is calculated based on how extreme the observed difference is relative to the permuted differences.
- **Bootstrap Confidence Interval**: This approach resamples each group (short and long) with replacement to generate new samples. For each bootstrap sample, it calculates the difference in mean `duration` between the "short" and "long" groups. The 2.5th and 97.5th percentiles of the resulting distribution of differences provide a 95% confidence interval, estimating the range within which the true difference in means likely falls.

(b) **Comparison to the Indicator Variable Model**:
- The indicator variable model uses a linear regression framework to estimate the mean difference between "short" and "long" groups, relying on assumptions of normality and equal variance of residuals.
- The permutation test, by contrast, is non-parametric and does not require these assumptions; it directly tests the null hypothesis by reshuffling labels.
- The bootstrap method is also non-parametric, generating a sampling distribution by resampling within each group. Unlike the indicator variable model, it provides a confidence interval rather than a direct p-value, allowing for an estimate of the plausible range for the mean difference without assuming a particular residual distribution.
