1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1. **Accuracy**  
   - **Scenario**: A spam email classifier.
   - **Rationale**: Accuracy is useful when the classes (spam or not spam) are balanced. It measures how well the classifier identifies both spam and non-spam emails correctly in relation to the total emails, making it suitable for balanced data where false positives and false negatives are not particularly costly.

2. **Sensitivity**  
   - **Scenario**: Medical screening for a serious disease (e.g., cancer detection).
   - **Rationale**: Sensitivity, or recall, is crucial when the goal is to identify as many actual positive cases as possible, such as in disease diagnosis. Missing a positive case (false negative) can be highly detrimental, so it's important to correctly identify most of the actual positives, even at the cost of some false positives.

3. **Specificity**  
   - **Scenario**: Fraud detection in financial transactions.
   - **Rationale**: Specificity is useful when the goal is to correctly identify negative cases, such as legitimate transactions that should not be flagged as fraudulent. In this case, false positives (incorrectly flagging a legitimate transaction) could lead to customer dissatisfaction and unnecessary investigations, so it's important to minimize these.

4. **Precision**  
   - **Scenario**: Search engine result ranking for a very specific query (e.g., searching for rare academic papers).
   - **Rationale**: Precision is important when the goal is to ensure that the positive results (documents or websites) returned by the search engine are highly relevant. It's better to show fewer, highly relevant results than to show many irrelevant ones, as users will be more satisfied with accurate search results, especially in specialized fields.

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Step 1: Create 80/20 split for training and testing data
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab, test_size=0.2, random_state=42)

# Step 2: Prepare the target variable (y) and feature matrix (X) for clf
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Target variable (hardcover or paperback)
X = ab_reduced_noNaN_train[['List Price']]  # Feature (List Price)

# Step 3: Train the DecisionTreeClassifier for clf with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Step 4: Prepare the feature matrix (X) for clf2
X2 = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]  # Features for clf2

# Step 5: Train the DecisionTreeClassifier for clf2 with max_depth=4
clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X2, y)

# Step 6: Make predictions on the test set
y_true = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # True values (hardcover or paperback)

# For clf, use only 'List Price' for predictions
X_test_clf = ab_reduced_noNaN_test[['List Price']]  # Test features for clf
y_pred_clf = clf.predict(X_test_clf)

# For clf2, use 'NumPages', 'Thick', and 'List Price' for predictions
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]  # Test features for clf2
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion Matrix for clf
cm_clf = confusion_matrix(y_true, y_pred_clf)
cm_clf2 = confusion_matrix(y_true, y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def calc_metrics(cm):
    TP = cm[1, 1]
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    
    sensitivity = TP / (TP + FN)
    specificity = TN / (TN + FP)
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    
    return np.round(sensitivity, 3), np.round(specificity, 3), np.round(accuracy, 3)

# Calculate metrics for both classifiers
sensitivity_clf, specificity_clf, accuracy_clf = calc_metrics(cm_clf)
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calc_metrics(cm_clf2)

# Report the results
print(f"Metrics for clf (DecisionTreeClassifier with List Price):")
print(f"Sensitivity: {sensitivity_clf}")
print(f"Specificity: {specificity_clf}")
print(f"Accuracy: {accuracy_clf}")

print("\nMetrics for clf2 (DecisionTreeClassifier with NumPages, Thick, List Price):")
print(f"Sensitivity: {sensitivity_clf2}")
print(f"Specificity: {specificity_clf2}")
print(f"Accuracy: {accuracy_clf2}")

# Visualize the confusion matrix for clf
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paperback', 'Hardcover'])
disp_clf.plot(cmap='Blues')
plt.title('Confusion Matrix for clf')
plt.show()

# Visualize the confusion matrix for clf2
disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paperback', 'Hardcover'])
disp_clf2.plot(cmap='Blues')
plt.title('Confusion Matrix for clf2')
plt.show()
````

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>

But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions.
